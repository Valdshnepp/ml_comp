```python
import numpy as np
import pandas as pd

nontext_path = 'data/sub_blend_best.csv'
text_path = 'new_tdata/sub_text_sgd_ens_fullfit_rank.csv'

sub_non = pd.read_csv(nontext_path)
sub_txt = pd.read_csv(text_path)

sub_non = sub_non.sort_values('id').reset_index(drop=True)
sub_txt = sub_txt.sort_values('id').reset_index(drop=True)

p_non = sub_non['is_fake'].values.astype(float)
p_txt = sub_txt['is_fake'].values.astype(float)

print('Loaded:', nontext_path, sub_non.shape)
print('Loaded:', text_path, sub_txt.shape)

```

```python

weights = [0.15, 0.2, 0.25,0.31, 0.35, 0.5, 0.65, 0.8, 0.9]

def to_rank01(x: np.ndarray) -> np.ndarray:
    order = x.argsort(kind='mergesort')
    ranks = np.empty_like(order, dtype=float)
    ranks[order] = np.linspace(0.0, 1.0, num=len(x), endpoint=True)
    return ranks

r_non = to_rank01(p_non)
r_txt = to_rank01(p_txt)

for w in weights:
    blend = w * r_txt + (1 - w) * r_non
    out = f"data/sub_blend_rank_text{int(w*100):02d}_non{int((1-w)*100):02d}.csv"
    pd.DataFrame({'id': sub_non['id'].values, 'is_fake': blend}).to_csv(out, index=False)
    print('saved', out)

```
